```python
from twitterscraper import query_tweets
import pandas as pd
import json
```

```python
def scrap(brand, qry, lang='ko'):
    #bd = pd.Timestamp('2014-01-01')
    #ed = pd.Timestamp('2018-12-31')

    tw = []
    for t in query_tweets(qry, lang=lang):#, begindate=bd, enddate=ed):
        tw.append({
            'id': t.id, 
            'date': t.timestamp, 
            'user': t.user, 
            'text': t.text, 
        })
    
    df = pd.DataFrame(tw).set_index('id')
    df.date = pd.to_datetime(df.date)
    df.to_pickle(brand + '.pkl')
    return df
```
