# DB 초기화

```python
with open('tokensfreq.json', encoding='UTF-8-sig') as f:
    wordfreqs = json.load(f)
    
def get_wordfreq(bname, topn=100, minfreq=5):
    try:
        wordfreq = {k:v for k,v in wordfreqs[bname].items() if v>=minfreq}
        return dict(sorted(wordfreq.items(), key=lambda x: x[1])[-topn:])
    except:
        return dict()
```

```python
df = pd.read_pickle('id_dict.pkl')
min_max_scaler = preprocessing.MinMaxScaler(feature_range=(0.1, 1))
X_train_minmax = min_max_scaler.fit_transform(df)
df[:] = X_train_minmax

def get_identity(bname):
    try:
        _df = df[[bname]].reset_index().rename(columns={'index':'key', bname:'value'})
        _df.value = (_df.value*100).astype(int)
        return _df.to_dict('record')
    except:
        return dict()
```

```python
def get_logo_fname(bname, where):
    try:
        return glob.glob(where + '/' + bname + '.*')[0].split('\\')[-1]
    except:
        return 'need file.jpg'
    
get_logo_fname('8seconds', 'uploads')
```

```python
db_init = []
ibrands = list(pd.read_excel('keywords and logos.xlsx', sheet_name='20190304').itertuples())[3:]
clusters = pd.read_excel('clustered.xlsx')[['bname','cluster']].set_index('bname')

for i, row in enumerate(ibrands):
    db_init.append({
        'model': 'app.brand', 
        'pk': i, 
        'fields': {
            'name': row.shortname, 
            'fullname_kr': row.koname, 
            'fullname_en': row.fullname, 
            'keywords': [w.strip() for w in row.keywords.split(',')], 
            'website': row.website,
            'origin': row.origin,
            'description': '',
            'history': '', 
            'logo': get_logo_fname(row.shortname, 'uploads'), 
            'cluster': str(clusters.cluster.get(row.shortname, '')), 
            'identity': json.dumps(get_identity(row.shortname), ensure_ascii=False), 
            'wordfreq': json.dumps(get_wordfreq(row.shortname), ensure_ascii=False), 
        }
    })
```

```python
with open('db_init.json', 'w') as f: #, encoding='UTF-8-sig'
    json.dump(db_init, f)#, ensure_ascii=False)
```
